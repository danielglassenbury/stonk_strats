# Yfinance

```python
import yfinance as yf
import pandas as pd
```

```python
# user inputs
ticker = "FPH.NZ"
PERIOD = "20y"
```

```python
stock = yf.Ticker(ticker)
print(stock.info)
print(stock.info.get("currency"))

data = stock.history(period=PERIOD, actions=True)
data['Dividends'] = data['Dividends'].fillna(0)
print(data.shape)
```

```python
data.head(-20)
```

```python
dividend_events = data[data['Dividends'] > 0]
dividend_events
```
